```python
# get url

import requests

url = input("type url: ")
requests.get(url)
```

```python
response = requests.get(url)

if response.status_code == 200:
    html_content = response.content
else:
    print("Failed to fetch the webpage.")
    exit()
```

```python
#parse

from bs4 import BeautifulSoup

soup = BeautifulSoup(html_content, "html.parser")
```

```python
new_soup = soup.prettify()
print(new_soup)
```

```python
keyword = input("search keyword: ")

names = soup.find_all(string=lambda text: keyword in text)
print(names)
```

```python
file_name = "Institute_of_Environmental_science_and_Geography" + ".txt"
with open (file_name , "w") as file:
    file.write("keyword: " + keyword + "\n")
    file.write("link: " + url + "\n" )
    for i in names:
        file.write(i + "\n")
```
